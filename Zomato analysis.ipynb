```python
df_country=pd.read_excel(r"C:\Users\nikhi\OneDrive\Desktop\5th SEM(Coding)\5-Days-Live-EDA-and-Feature-Engineering-main\Country-Code.xlsx")
df_country.head(5)
```

```python
df_country.columns
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how="left")
final_df.head(2)
```

```python
final_df.dtypes
```

```python
country_names=final_df.Country.value_counts().index
country_names
```

```python
country_values=final_df.Country.value_counts().values
country_values
```

```python
plt.pie(country_values[:3],labels=country_names[:3],autopct="%1.2f%%")
```

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
ratings
```

```python
sns.barplot(x='Aggregate rating',y='Rating color',hue='Rating color',data=ratings,palette=['White','Red','Orange','green','green'])
```

```python
sns.countplot(x='Rating color',data=ratings,palette=['White','Red','Orange','green','green'])
```
